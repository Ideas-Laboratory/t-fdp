```python
import pandas as pd
import numpy as np
from seaborn import heatmap
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'text.usetex':False, 'font.family':'sans-serif','figure.facecolor':"white",'axes.facecolor':"white"})
```

```python
df_np1_tsnet = pd.read_csv("./results/tsNET_NP1.csv",index_col=0)
df_np2_tsnet = pd.read_csv("./results/tsNET_NP2.csv",index_col=0)

plt.figure(figsize=(10,8))
plt.figure(figsize=(6,4))
ax = plt.gca()
plt.plot(df_np1_tsnet.mean()[:-1],marker='^',markersize=10,c='purple',markerfacecolor="white")
plt.plot(df_np2_tsnet.mean()[:-1],marker='s',markersize=10,c='g',markerfacecolor="white")
plt.plot(df_np1_tsnet.mean()[-1:],marker='^',markersize=10,c='purple',markerfacecolor="white")
plt.plot(df_np2_tsnet.mean()[-1:],marker='s',markersize=10,c='g',markerfacecolor="white")
yticks = 0.36 + np.array(np.arange(11)) * 0.09
ax.set_yticks(yticks)
plt.ylim(0.36,0.72)
plt.legend(["NP1","NP2"])

plt.yticks(fontsize=15,fontname="Times New Roman")
plt.xticks(fontsize=15,fontname="Times New Roman")
plt.savefig("Figs/SuppFig12Diffpara/tsNET.svg")
plt.show()
```

```python
df_np1_drgraph = pd.read_csv("./results/DRGraph_NP1.csv",index_col=0)
df_np2_drgraph = pd.read_csv("./results/DRGraph_NP2.csv",index_col=0)

plt.figure(figsize=(10,8))
plt.figure(figsize=(6,4))
ax = plt.gca()
plt.plot(df_np1_drgraph.mean()[:-1],marker='^',markersize=10,c='purple',markerfacecolor="white")
plt.plot(df_np2_drgraph.mean()[:-1],marker='s',markersize=10,c='g',markerfacecolor="white")
plt.plot(df_np1_drgraph.mean()[-1:],marker='^',markersize=10,c='purple',markerfacecolor="white")
plt.plot(df_np2_drgraph.mean()[-1:],marker='s',markersize=10,c='g',markerfacecolor="white")
yticks = 0.20 + np.array(np.arange(12)) * 0.12
ax.set_yticks(yticks)
plt.ylim(0.20,0.68)
plt.legend(["NP1","NP2"])
plt.yticks(fontsize=15,fontname="Times New Roman")
plt.xticks(fontsize=15,fontname="Times New Roman")
plt.savefig("Figs/SuppFig12Diffpara/DRGraph.svg")
plt.show()
```
