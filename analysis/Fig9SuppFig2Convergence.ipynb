```python
import numpy as np
import pandas as pd
```

```python
df_start_value = pd.DataFrame([0.060311902921566435, 0.17803295231325283, 0.10216890004338175]).T
df_start_value.columns=['NP1','NP2','SE']
```

```python
dfExact = pd.concat([df_start_value,pd.read_csv("./results/Convergence/Exact.csv")], axis=0 ,ignore_index=True)
dfBH =  pd.concat([df_start_value,pd.read_csv("./results/Convergence/BH.csv")], axis=0 ,ignore_index=True)
dfRVS = pd.concat([df_start_value,pd.read_csv("./results/Convergence/RVS.csv")], axis=0 ,ignore_index=True)
dfibFFT_CPU = pd.concat([df_start_value,pd.read_csv("./results/Convergence/ibFFT_CPU.csv")], axis=0 ,ignore_index=True)
dfibFFT_GPU = pd.concat([df_start_value,pd.read_csv("./results/Convergence/ibFFT_GPU.csv")], axis=0 ,ignore_index=True)
```

```python
from seaborn import heatmap
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'text.usetex':False, 'font.family':'sans-serif','figure.facecolor':"white",'axes.facecolor':"white"})

colors =  ['#1F77B4', '#FF7F0E', '#A7A7A7', '#2CA02C', '#D62728']
```

```python
x = list(np.arange(60) * 5)
x.append(300)
x2 = list(np.arange(100) * 3) 
x2.append(300)
```

```python
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams.update(mpl.rcParamsDefault)

from scipy.interpolate import make_interp_spline

# mpl.rcParams['text.usetex'] = True

plt.figure(figsize=(12,7))
plt.grid(True)

# plt.xscale("log")
ax = plt.gca()


ax.axis("on")
# ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.spines["right"].set_visible(True)
ax.spines["top"].set_visible(True)
ax.spines["bottom"].set_visible(True)
ax.spines["left"].set_visible(True)

ax.spines["right"].set_color("#AAAAAA")
ax.spines["top"].set_color("#AAAAAA")
ax.spines["bottom"].set_color("#AAAAAA")
ax.spines["left"].set_color("#AAAAAA")
# ax.set_xlim(-5,305)

lw = 3

ax.plot(dfExact["SE"].loc[x2],zorder=2,lw=lw,c=colors[0],markerfacecolor="white")
ax.plot(dfBH["SE"].loc[x2],zorder=2,lw=lw,c=colors[1],markerfacecolor="white")
ax.plot(dfRVS["SE"].loc[x2],zorder=2,lw=lw,c=colors[2],markerfacecolor="white")
ax.plot(dfibFFT_CPU["SE"].loc[x2],zorder=2,lw=lw,c=colors[3],markerfacecolor="white")
ax.plot(dfibFFT_GPU["SE"].loc[x2],zorder=2,lw=lw,c=colors[4],markerfacecolor="white")

ax.set_ylim(0.0770,0.107)
# ax.set_yticks([0.077,0.080,0.083,0.086,0.089,0.092,0.095,0.098,0.101,0.104,0.107])
ax.set_yticks([0.075,0.080,0.085,0.090,0.095,0.100,0.105,0.110])


plt.yticks(fontsize=25,fontname="Times New Roman")
plt.xticks(fontsize=25,fontname="Times New Roman")
ax2 = ax.twinx()
lw = 3

ax2.plot(dfExact["NP1"].loc[x],ls="--",lw=lw,zorder=2,c=colors[0],markerfacecolor="white")
ax2.plot(dfBH["NP1"].loc[x],ls="--",lw=lw,zorder=2,c=colors[1],markerfacecolor="white")
ax2.plot(dfRVS["NP1"].loc[x],ls="--",lw=lw,zorder=2,c=colors[2],markerfacecolor="white")
ax2.plot(dfibFFT_CPU["NP1"].loc[x],ls="--",lw=lw,zorder=2,c=colors[3],markerfacecolor="white")
ax2.plot(dfibFFT_GPU["NP1"].loc[x],ls="--",lw=lw,zorder=2,c=colors[4],markerfacecolor="white")


ax2.set_ylim(0,0.70)

plt.yticks(fontsize=25,fontname="Times New Roman")
plt.xticks(fontsize=25,fontname="Times New Roman")
plt.savefig("./Figs/Fig9SuppFig2Convergence/per_iteration.svg")
plt.show()
```
